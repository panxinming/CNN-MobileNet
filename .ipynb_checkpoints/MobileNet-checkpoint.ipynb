# Mobile Net For Image Classification

```python
import tensorflow as tf
import matplotlib.pyplot as plt
print(tf.__version__)
```

## 1. Load Fashion Mnist Data

```python
#Load data from the Fashion MNIST dataset
fashion_mnist = tf.keras.datasets.fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalization (归一化)
x_train, x_test = x_train / 255.0, x_test / 255.0
```

## 2. Visualization

```python
#Visualizing Fashion-MNIST data. We randomly choose 25 images from the train dataset.
label_map = [
    't-shirt/top', 'trouser', 'pullover', 'dress', 'coat', 
    'sandal', 'shirt', 'sneaker', 'bag', 'ankle boot'
]

fig, ax = plt.subplots(5, 5, figsize=(8, 8))
for j in range(5):
    for k in range(5):
        i = np.random.choice(range(len(x_train)))
        ax[j][k].set_axis_off()
        ax[j][k].imshow(x_train[i:i+1][0], cmap='gray')
        ax[j][k].set_title(label_map[y_train[i]])
```

## 3. Tensorflow Basic CNN Model

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(32, (3,3),strides=1,padding="same", input_shape=(28, 28, 1)),
                                    tf.keras.layers.MaxPooling2D((2, 2)),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(10, activation="softmax")])

model.compile(optimizer=tf.keras.optimizers.Nadam(0.001),
              loss=tf.keras.metrics.sparse_categorical_crossentropy,
              metrics=['accuracy'])

model.fit(x=x_train, 
          y=y_train,
          batch_size=256,
          epochs=10, 
          validation_data=(x_test, y_test))
```

## 4. Tensorflow Basic CNN with Mobile Net

```python
i = tf.keras.Input(shape=(28, 28, 1))
x = tf.keras.applications.mobilenet_v2.preprocess_input(i)
preprocessor = tf.keras.Model(inputs = [i], outputs = [x])
```

```python
model = tf.keras.models.Sequential([preprocessor,
                                    tf.keras.layers.Conv2D(32, (3,3),strides=1,padding="same", input_shape=(28, 28, 1)),
                                    tf.keras.layers.MaxPooling2D((2, 2)),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(10, activation="softmax")])

model.compile(optimizer=tf.keras.optimizers.Nadam(0.001),
              loss=tf.keras.metrics.sparse_categorical_crossentropy,
              metrics=['accuracy'])

model.fit(x=x_train, 
          y=y_train,
          batch_size=256,
          epochs=10, 
          validation_data=(x_test, y_test))
```
